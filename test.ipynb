```python
model_name = 'final_model'
model = load_model(model_name)
```

```python
def predict(X):
  X_new = resize(images=X, size=[299,299])
  X_new = X_new / 255.0
  predictions = model.predict(x=X_new, batch_size=X.shape[0])
  pre_ohe = to_categorical(np.argmax(predictions, axis=1), 3)
  return convert_to_tensor(pre_ohe)
```

```python
def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]

val_data_dir = 'data/val/'

def create_true_array():
  true_arr = np.ndarray((0,3))
  for i in range(3):
    if i==0:
      for j in range(360):
        true_arr=np.append(true_arr,[[1., 0., 0.]],0)
    elif i==1:
      for j in range(360):
        true_arr=np.append(true_arr,[[0., 1., 0.]],0)
    elif i==2:
      for j in range(360):
        true_arr=np.append(true_arr,[[0., 0., 1.]],0)
  return true_arr

l = []
p = []
true_arr = create_true_array()
c = 0
dim = 250
z=sorted(os.listdir(val_data_dir), key=natural_keys)
for x in z:
  w=sorted(os.listdir(val_data_dir + x), key=natural_keys)
  for y in w:
    p.append(y)
    img = load_img(val_data_dir+x+'/'+y)
    pic = img_to_array(img)
    fig = resize(images=pic, size=[dim, dim])
    l.append(fig)
    if len(l) == 100 or (x==z[-1] and y==w[-1]):
      tot = np.stack(l, 0)
      tensor = convert_to_tensor(l)
      #print(tot.shape)
      result = predict(tensor)
      for i in range(len(l)):
        if result.numpy().tolist()[i] != true_arr.tolist()[i]:
          print(p[i])
          print("predicted: "+str(result.numpy().tolist()[i]))
          print("real: "+str(true_arr.tolist()[i])+'\n')
          c += 1
      true_arr=true_arr[len(l):]
      l.clear()
      p.clear()
true_arr = create_true_array()
```
